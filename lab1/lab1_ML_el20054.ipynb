3.α.  Θα χρησιμοποιήσετε όλα τα χαρακτηριστικά του συνόλου δεδομένων για την εκπαίδευση των ταξινομητών ή θα επιλέξετε κάποια από αυτά;  Θα κάνετε κάποια συνένωση κάποιων χαρακτηριστικών για δημιουργία νέων χαρακτηριστικών για το μοντέλο σας;


Μπορούμε να μην χρησιμοποιήσουμε όλα τα χαρακτηριστικά και να ενοποιήσοουμε εκείνα που έχουν τιμές στον correlation πίνακα κοντά στο 1. Τέτοια χαρακτηριστικά είναι τα: (MaxTemp - MinTemp - Temp9am - Temp3pm = Temperature) , (Pressure9am - Pressure3pm = Presssure) και (WindGustSpeed - WindSpeed9am - WindSpeed3pm = WindSpeed). Επίσης θα κάνουμε drop την στήλη Id καθώς δεν είναι χρήσιμη για την εκπάιδευση

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create a new column 'Temperature' by computing average value of 4 Temperature columns
data['Temperature'] = (data['MaxTemp'] + data['Temp3pm'] + data['Temp9am'] + data['MinTemp'])/4

# Create a new column 'WindSpeed' by computing average value of 3 WindSpeed columns
data['WindSpeed'] = (data['WindGustSpeed'] + data['WindSpeed9am'] + data['WindSpeed3pm'])/3

# Create a new column 'Pressure' by computing average value of 2 Pressure columns
data['Pressure'] = (data['Pressure9am'] + data['Pressure3pm'])/2

# Drop the previous temperature-related columns
data.drop(['MaxTemp', 'Temp3pm', 'Temp9am', 'MinTemp','WindGustSpeed', 'WindSpeed9am', 'WindSpeed3pm', 'Pressure9am', 'Pressure3pm', 'id'], axis=1, inplace=True)

# Display the updated DataFrame
print(data.head())
```

3.β. Υπάρχουν απουσιάζουσες τιμές; Γράψτε κατάλληλο κώδικα ώστε να χειριστείτε αυτές τις τιμές.

```python
# Show missing values
data.isnull().sum()
```

Παρατηρούμε λοιπόν πως έχουμε απουσιάζουσες τιμές στα δεδομένα μας, όπως θα αναμέναμε. Θα μπορούσαμε να διαγράψουμε τα δεδομένα αυτά από το dataset αλλά επειδή είναι πολλά και δεν θέλουμε να διαγράψουμε τόσα πολλά δεοδομένα εκπίδευσης θα χρησιμοποιήσουμε τον μετασχηματιστή Imputer που αντικαθιστά τις απουσιάζουσες τιμές με τηη μέση τιμή ή την πιο συχνή τιμή του χαρακτηριστικού

```python
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder

# Columns with missing values
columns_with_missing_values = data.columns[data.isnull().any()]

# Separate columns into numeric and non-numeric types
numeric_columns = data[columns_with_missing_values].select_dtypes(include=['float64', 'int64']).columns
non_numeric_columns =data[columns_with_missing_values].select_dtypes(exclude=['float64', 'int64']).columns

# Instantiate the SimpleImputer with the 'mean' strategy for numeric columns
numeric_imputer = SimpleImputer(strategy='mean')
data[numeric_columns] = numeric_imputer.fit_transform(data[numeric_columns])

# Instantiate the SimpleImputer with the 'most_frequent' strategy for non-numeric columns
non_numeric_imputer = SimpleImputer(strategy='most_frequent')
data[non_numeric_columns] = non_numeric_imputer.fit_transform(data[non_numeric_columns])

# Display the DataFrame with filled values
print(data.head())
```

3.γ. Γράψτε κώδικα για την κατάλληλη μετατροπή των κατηγορικών μεταβλητών ώστε να μπορούν να τους διαχειριστούν οι ταξινομητές που θα χρησιμοποιήσετε.

Οι κατηγορικές στήλες είναι οι: Date, Loation, WindGustDir, WindDir9am, WindDir3pm

```python
# handle categorical features
#Firstly we proccess column 'Date'

# Convert 'Date' to datetime format
data['Date'] = pd.to_datetime(data['Date'])

# Extract year, month, and day
data['Year'] = data['Date'].dt.year
data['Month'] = data['Date'].dt.month
data['Day'] = data['Date'].dt.day

# Drop the original 'Date' column
data.drop('Date', axis=1, inplace=True)

print(data.head())
```

```python
#We then process column Location
from sklearn.preprocessing import LabelEncoder

# Initialize the LabelEncoder
label_encoder = LabelEncoder()

# Encode 'Location' to numerical values
data['Location_Code'] = label_encoder.fit_transform(data['Location'])

# Drop the original 'Location' column
data.drop('Location', axis=1, inplace=True)
print(data.head())
```

```python
# Create a mapping for Wind Direction
wind_direction_mapping = {'W': 0, 'N': 1, 'WSW': 2, 'SW': 3, 'NW': 4, 'WNW': 5, 'E': 6, 'S': 7,
                          'SSW': 8, 'SE': 9, 'SSE': 10, 'ENE': 11, 'NNW': 12, 'ESE': 13, 'NE': 14, 'NNE': 15}

# Apply the mapping to each Wind Direction column
wind_direction_columns = ['WindGustDir', 'WindDir9am', 'WindDir3pm']
for column in wind_direction_columns:
    data[column] = data[column].map(wind_direction_mapping)
print(data.head())
```
